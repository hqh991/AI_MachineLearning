```python
import numpy as np
import pandas as pd 
url='50_Startups.csv'
data = pd.read_csv(url)
from matplotlib import pyplot as plt
data.hist() 
plt.show
```

```python
correlation = data.corr() 
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
```

```python
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Profit']),y=data['R&D Spend'])
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X=pd.DataFrame(data['Profit']),y=data['R&D Spend'])
#do chinh xac
```

```python
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize = (9,9),color='black',xlim = (0,200000))
y_pred = model.predict(X=pd.DataFrame(data['Profit']))
plt.plot(data['Profit'],y_pred,color='blue')
```

```python
model= linear_model.LinearRegression()
X=pd.DataFrame([data['Profit'],data['Profit']**2,data['Profit']**3,data['Profit']**4]).T
model.fit(X,y= data['R&D Spend'])
print('INTERCEPT cua mo hinh:',model.intercept_)
print('COEFFIENT cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['R&D Spend']))
```

```python
range = np.arange(1.5,200000,10000)
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize = (9,9),color='black',xlim = (0,200000))
predictor = pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
