```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np


url = 'Position_Salaries.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr() 
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
sns.pairplot(data)
```

```python
data.plot(kind='scatter',x='Salary',y='Level',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()

X=pd.DataFrame(data['Salary'])
y=data['Level']

model.fit(X,y)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
X=pd.DataFrame(data['Salary'])
y=data['Level']
model.fit(X,y)
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X,y)
```

```python
data.plot(kind='scatter',x='Salary',y='Level',figsize=(9,9),color = 'black',xlim=(0.7))
y_pred = model.predict(X)
plt.plot(X,y_pred,color='blue')
#Bac 1
```

```python
#Bac 3
X=pd.DataFrame([data['Salary'],data['Salary']**2,data['Salary']**3]).T
y=data['Level']
model.fit(X,y)
print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['Level']))
```

```python
data.plot(kind='scatter',x='Salary',y='Level',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['Salary'].min(),data['Salary'].max(),10) 
predictor = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
