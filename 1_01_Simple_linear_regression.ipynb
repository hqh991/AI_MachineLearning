```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np


url = '1.01. Simple linear regression.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr() 
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
sns.pairplot(data)
```

```python
data.plot(kind='scatter',x='GPA',y='SAT',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()

X=pd.DataFrame(data['GPA'])
y=data['SAT']

model.fit(X,y)

print(model.intercept_)
print(model.coef_)
model.score(X,y)
#bac 1
```

```python
data.plot(kind='scatter',x='GPA',y='SAT',figsize=(9,9),color = 'black',xlim=(0.5))
y_pred = model.predict(X)
plt.plot(X,y_pred,color='blue')
```

```python
#Bac 2
X=pd.DataFrame([data['GPA'],data['GPA']**2]).T
y=data['SAT']
model.fit(X,y)
print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['SAT']))
```

```python
data.plot(kind='scatter',x='GPA',y='SAT',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['GPA'].min(),data['GPA'].max(),0.1) 
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```

```python
#Bac 3
X=pd.DataFrame([data['GPA'],data['GPA']**2,data['GPA']**3]).T
y=data['SAT']
model.fit(X,y)
print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['SAT']))
```

```python
data.plot(kind='scatter',x='GPA',y='SAT',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['GPA'].min(),data['GPA'].max(),0.1) 
predictor = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
