```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np


url = 'data_crab.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
```

```python
data.plot(kind = 'scatter', x = 'CL', y = 'CW',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
```

```python
X=pd.DataFrame(data['CL'])
y=data['CW']

model.fit(X,y)

print(model.intercept_)
print(model.coef_)
model.score(X,y)
```

```python
data.plot(kind='scatter',x='CL',y='CW',figsize=(9,9),color = 'black',xlim=(0.5))
y_pred = model.predict(X)
plt.plot(X,y_pred,color='blue')
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np


url = 'passenger.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
     
```

```python
data.plot(kind = 'scatter', x = 'Dollars (Y)', y = 'Miles(X)',figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
```

```python
X=pd.DataFrame(data['Dollars (Y)'])
y=data['Miles(X)']

model.fit(X,y)

data.plot(kind='scatter',x='Dollars (Y)',y='Miles(X)',figsize=(9,9),color = 'black',xlim=(0.5))
y_pred = model.predict(X)
plt.plot(X,y_pred,color='blue')





print(model.intercept_)
print(model.coef_)
model.score(X,y)
```

```python
X=pd.DataFrame([data['Dollars (Y)'],data['Dollars (Y)']**2,data['Dollars (Y)']**3,data['Dollars (Y)']**4,data['Dollars (Y)']**5,
                data['Dollars (Y)']**6,data['Dollars (Y)']**7,data['Dollars (Y)']**8,data['Dollars (Y)']**9,data['Dollars (Y)']**10]).T
y=data['Miles(X)']

model.fit(X,y)



print(model.intercept_)
print(model.coef_)
model.score(X,y)
```

```python

data.plot(kind='scatter',x='Dollars (Y)',y='Miles(X)',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['Dollars (Y)'].min(),data['Dollars (Y)'].max(),1) 
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
