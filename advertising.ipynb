```python
import numpy as np
import pandas as pd 
url='Advertising.csv'
data = pd.read_csv(url)
from matplotlib import pyplot as plt
data.hist() 
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
```

```python
data.plot(kind='scatter',x='sales',y='TV',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['sales']),y=data['TV']) #ham day hoc
print(model.intercept_)
print(model.coef_)
```

```python
range = np.arange(1.5,30,0.1)
data.plot(kind='scatter',x='sales',y='TV',figsize = (9,9),color='black',xlim = (0,40))
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```

```python
model= linear_model.LinearRegression()
X=pd.DataFrame([data['sales'],data['sales']**2,data['sales']**3,data['sales']**4,data['sales']**5]).T
model.fit(X,y= data['TV'])
print('INTERCEPT cua mo hinh:',model.intercept_)
print('COEFFIENT cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['TV']))
```
